## Stemming

```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

## Porter Stemmer

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for word in words:
    print(f"{word} ----> {ps.stem(word)}")
```

## RegExpStemmer

```python
from nltk.stem import RegexpStemmer

rs = RegexpStemmer('ing$|s$|e$|able$', min=4)

for word in words:
    print(f"{word} ----> {rs.stem(word)}")
```

```python
rs.stem("Baking")
```

## Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')
for word in words:
    print(f"{word} ----> {ss.stem(word)}")
```

```python
ss.stem("Justification")
```

## Lemmatization

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()

for word in words:
    print(f"{word} ----> {wl.lemmatize(word,pos='v')}")
```

## Bag Of words Example

```python
import pandas as pd
messages=pd.read_csv('spam.csv', encoding='latin-1')
```

```python
# import os
# os.getcwd()
```

```python
messages.head()
```

```python
messages = messages[['v1','v2']]
```

```python
messages = messages.rename(columns={'v1':'label', 'v2':'message'})
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
## Remove all the stopword and then store

corpus = []
len(messages)

for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    
    review=review.split()
    print(review)

    review = [wl.lemmatize(word) for word in review if not word in stopwords.words('english')]
    print(review)
    review =' '.join(review)
    print(review)

    corpus.append(review)
```

```python
# pip install scikit-learn
```

```python
# Create a BOW model

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
cv.vocabulary_
```

## N Grams

```python
# Create a BOW model

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500,binary=True,ngram_range=(3,3))
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
cv.vocabulary_
```
